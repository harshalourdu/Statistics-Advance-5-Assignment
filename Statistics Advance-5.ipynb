```python
# 1. Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

import scipy.stats as stats
import math

# Given values
mean = 50
std_dev = 5
n = 30  # assumed sample size for the calculation
confidence_level = 0.95

# Z-score for 95% confidence
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Standard error
standard_error = std_dev / math.sqrt(n)

# Confidence interval
margin_of_error = z_score * standard_error
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

confidence_interval
```

```python
# Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

# Given values
smokers = 60
total_population = 500
confidence_level = 0.95

# Sample proportion
p_hat = smokers / total_population

# Z-score for 95% confidence
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Standard error for proportion
standard_error = math.sqrt(p_hat * (1 - p_hat) / total_population)

# Confidence interval
margin_of_error = z_score * standard_error
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

confidence_interval
```

```python
# Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

mean = 75
std_dev = 12
n = 30  # assumed sample size

# Z-score for 90% confidence
z_score = stats.norm.ppf(1 - (1 - 0.90) / 2)

# Standard error
standard_error = std_dev / math.sqrt(n)

# Confidence interval
margin_of_error = z_score * standard_error
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

confidence_interval
```

```python
# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

import matplotlib.pyplot as plt
import numpy as np

# Define degrees of freedom and chi-square statistic
df = 10
chi_square_value = 15

# Define range for the plot
x = np.linspace(0, 30, 1000)
y = stats.chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, y, label=f'Chi-Square Distribution (df={df})')

# Shade the area for chi-square statistic of 15
plt.fill_between(x, y, where=(x >= chi_square_value), color='skyblue', alpha=0.5, label='Area for Chi-Square > 15')

# Labels and title
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-Square Value')
plt.ylabel('Density')
plt.legend()

plt.show()
```

```python
# Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

coke_preference = 520
sample_size = 1000
confidence_level = 0.99

# Sample proportion
p_hat = coke_preference / sample_size

# Z-score for 99% confidence
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Standard error for proportion
standard_error = math.sqrt(p_hat * (1 - p_hat) / sample_size)

# Confidence interval
margin_of_error = z_score * standard_error
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

confidence_interval
```
